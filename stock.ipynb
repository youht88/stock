```python
import pandas as pd
import csv

for i in range(1,178):  # 爬取全部177页数据
    url = 'http://s.askci.com/stock/a/?reportTime=2018-12-31&pageNum=%s' % (str(i))
    tb = pd.read_html(url)[3] #经观察发现所需表格是网页中第4个表格，故为[3]
    tb.to_csv(r'1.csv', mode='a', encoding='utf_8_sig', header=1, index=0)
    print('第'+str(i)+'页抓取完成')
```

```python
import pandas as pd
from pyecharts.charts import Bar,Line
tb=pd.read_csv('1.csv')
#print(tb.head())
tb.columns=["no","gpdm","gpmc","name","cs0000","qx0000","zysr","jlr","ygrs","ssrq","zgs",\
                             "gscb","hyfl","cplx","zyyw"]
tb=tb.query("name!='公司名称'")
tb=tb.drop_duplicates()
tb["zysr"]=tb["zysr"].str.extract(r'(\d+\.?\d*)' ,expand=False).apply(pd.to_numeric)
#tb["zysr1"]=tb["zysr"].str.slice(0,-1)
tb["jlr"]=tb["jlr"].str.extract(r'(\d+\.?\d*)',expand=False).apply(pd.to_numeric)
tb["ygrs"]=tb["ygrs"].apply(pd.to_numeric)
#tb["ssrq"]=tb["ssrq"].apply(pd.to_datetime,"%Y-%M-%d")
tb=tb.loc[:,["name","gpdm","gpmc","cs0000",\
             "qx0000","hyfl","zysr","jlr","ygrs","ssrq"]].query("zysr>1000" )
tb=tb.sort_values(['ssrq','ygrs'],ascending=True)
#tb=tb.groupby(["cs0000","qx0000"]).sum()[["zysr","jlr","ygrs"]]
tb=tb.groupby("hyfl").sum()[["zysr","jlr","ygrs"]]
tb
#tb.groupBy("hyfl").agg(Map("zysr" -> "sum", "jlr" -> "sum"))
#.toDF("hyfl","sumZysr","sumJlr","cnt").show
```

```python
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.globals import ThemeType

bar=Bar(init_opts=opts.InitOpts(theme=ThemeType.WESTEROS,width="800px",height="4200px"))
bar.add_xaxis(list(tb.hyfl.to_numpy()))
#bar.add_yaxis("员工人数",list(tb.ygrs.to_numpy().astype('float')))
bar.add_yaxis("净利润",list(tb.jlr.to_numpy()))
bar.add_yaxis("主营收入",list(tb.zysr.to_numpy()))
bar.reversal_axis()
bar.set_global_opts(title_opts=opts.TitleOpts(title="上市公司信息",subtitle="2018-12-31"))
bar.render_notebook()
```
